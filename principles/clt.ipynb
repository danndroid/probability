```python
import numpy as np
from scipy.stats import bernoulli, norm, poisson, uniform

import matplotlib.pyplot as plt
plt.style.use('bmh')
import seaborn as sns
```

```python
Unirorm = uniform(0,1)
```

```python
experiments = 25

samples = 50
sample_size = 50 #### n

step = 10

experiment_size = []
meta_mean = []
experiment_means = []
for i in range(experiments):
    experiment_means = []
    for j in range(samples):
        #print(i,j,sample_size)
        sample = Unirorm.rvs(sample_size)
        m = np.mean(sample)
        experiment_means.append(m)

    experiment_size.append(sample_size)
    sample_size = sample_size + step
    meta_mean.append(experiment_means)
    #print('')
meta_mean = np.array(meta_mean)
```

```python
fig = plt.figure(figsize=(17,15))

for i , experiment in enumerate(meta_mean):
    #experiment = experiment_means[:exp]
    m, v, s = np.mean(experiment), np.std(experiment), np.var(experiment)
    
    ax = fig.add_subplot(5, 5, i+1)
    ax.set_title(f'N={experiment_size[i]}: m={m:.4f}-v={v:.4f}', fontsize=15)
    sns.histplot(experiment, bins=30, kde=True, color='gray', ax=ax)
    ax.lines[0].set_color('blue')
    ax.set_xlim(0,1)

    #plt.legend()
plt.suptitle('Distributions', fontsize=20)
plt.tight_layout()
plt.show()
```
