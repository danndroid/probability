```python
import numpy as np
from scipy.stats import bernoulli, norm, poisson, uniform
from sklearn.metrics import mutual_info_score

import matplotlib.pyplot as plt
plt.style.use('bmh')
import seaborn as sns
```

```python
Uniform = uniform(0,1)
```

```python
Nx = Uniform.rvs(1000)
Ny = Uniform.rvs(1000)

Y = 0.5*Nx + Ny
```

```python
fig = plt.figure(figsize=(8,3))

ax = fig.add_subplot(131, title='P(X)')
ax.hist(Nx, 30, density=True, alpha=0.75)
ax.set_xlabel('X')

ax = fig.add_subplot(132, title='P(Y)')
ax.hist(Ny, 30, density=True, alpha=0.75)
ax.set_xlabel('Y')

ax = fig.add_subplot(133, title='P(Y)')
ax.hist(Y, 30, density=True, alpha=0.75)
ax.set_xlabel('Y')

plt.tight_layout()
plt.show()
```

```python
H, yedges, xedges = np.histogram2d(Y, Nx, bins=30)
```

```python
plt.pcolormesh(xedges, yedges, H)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
