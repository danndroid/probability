```python
import numpy as np
from scipy.stats import bernoulli, norm, poisson
from sklearn.metrics import mutual_info_score

import matplotlib.pyplot as plt
plt.style.use('bmh')
import seaborn as sns
```

### Uniform

```python
x = np.random.uniform(0, 1, 100)
plt.hist(x)
plt.show()
```

### Normal

```python
m = 0
s = 1
standard = norm(m,s)
x = standard.rvs(100)
plt.hist(x)
plt.show()
```

### Bernoulli

```python
p = 0.3
Bern = bernoulli(p)
x = Bern.rvs(100)
plt.hist(x)
plt.show()
```

### Poisson

```python
mu = 0.6
Bern = poisson(mu)
x = Bern.rvs(1000)
plt.hist(x)
plt.show()
```
